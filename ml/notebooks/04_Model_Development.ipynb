# Module 5 - Model Development

## Purpose

This notebook trains and evaluates the core classification models for flood prediction using the preprocessed data from Module 4.

## Theory

Decision trees, random forests, k-nearest neighbors, and boosting-based models each learn different decision boundaries and provide complementary strengths for classification tasks.

## Advantages

These models provide interpretable baselines, robust ensemble performance, and flexible nonlinear decision boundaries.

## Limitations

The dataset is small and imbalanced, so some metrics may be sensitive to class distribution and threshold choices.

## Workflow

1. Load the processed data and split it into train and test sets.
2. Train each model independently and evaluate its performance.
3. Generate confusion matrices, ROC curves, and feature-importance plots.
4. Summarize metrics and save the report and artifacts.

## Applications

These trained models provide the foundation for model comparison, selection, and later deployment-ready serialization.

```python
from utils.model_training import load_training_data

X_train, X_test, y_train, y_test = load_training_data()
print('Train shape:', X_train.shape)
print('Test shape:', X_test.shape)
print('Target distribution:', y_train.value_counts().to_dict())
```

## Decision Tree

### Purpose

Train and evaluate a simple interpretable tree-based model.

### Theory

Decision trees split the feature space recursively to maximize purity and are easy to inspect.

### Advantages

They are simple, fast, and highly interpretable.

### Limitations

They can overfit the training data when the tree becomes too deep.

### Workflow

Fit the model, generate predictions and probabilities, and compute the evaluation metrics.

### Applications

Decision trees are useful as early baselines and explainable models.

```python
from utils.model_training import decision_tree_model

tree_model, tree_predictions, tree_metrics = decision_tree_model(X_train, X_test, y_train, y_test)
tree_metrics
```

## Random Forest

### Purpose

Train a robust ensemble of trees to improve stability and generalization.

### Theory

Random forests average the predictions of many trees and reduce overfitting compared with a single tree.

### Advantages

They usually provide better performance and better resistance to noise.

### Limitations

They are slower and less interpretable than single trees.

### Workflow

Fit the forest, collect predictions and probabilities, and compute the metrics.

### Applications

Random forests are a reliable benchmark for structured tabular data.

```python
from utils.model_training import random_forest_model

forest_model, forest_predictions, forest_metrics = random_forest_model(X_train, X_test, y_train, y_test)
forest_metrics
```

## K-Nearest Neighbors

### Purpose

Train a distance-based classifier to compare a non-parametric baseline.

### Theory

KNN classifies a sample based on the labels of its nearest neighbors in feature space.

### Advantages

It is simple and works well when the feature space is well structured.

### Limitations

It scales poorly with larger datasets and is sensitive to feature scaling.

### Workflow

Fit the model, generate predictions and probabilities, and evaluate the outputs.

### Applications

KNN serves as an interpretable local-method baseline.

```python
from utils.model_training import knn_model

knn_model_obj, knn_predictions, knn_metrics = knn_model(X_train, X_test, y_train, y_test)
knn_metrics
```

## XGBoost

### Purpose

Train a boosting-based classifier that often delivers strong performance on tabular data.

### Theory

Boosting ensembles build weak learners iteratively to reduce bias and improve predictive power.

### Advantages

It often performs well on structured data and offers robust predictive accuracy.

### Limitations

It can be more sensitive to tuning and requires careful validation.

### Workflow

Fit the model, generate predictions and probabilities, and compute the evaluation metrics.

### Applications

XGBoost is widely used in production-grade tabular classification tasks.

```python
from utils.model_training import xgboost_model

xgb_model, xgb_predictions, xgb_metrics = xgboost_model(X_train, X_test, y_train, y_test)
xgb_metrics
```

## Evaluation Summary and Reporting

### Theory

### Advantages

It provides a compact view of model quality and a durable artifact for the next module.

### Limitations

The summary is only as informative as the metrics and validation split used.

### Workflow

Aggregate the results into a DataFrame, write the markdown report, and save the figures.

### Applications

The summary supports model comparison and later selection in the next module.
