# Module 2: Data Collection and Data Understanding

This notebook focuses only on loading, validating, and understanding the provided datasets for the Rising Waters project. It does not perform visualization, preprocessing, or modeling.

## Step 1: Import Libraries

- `numpy`: Numerical operations and array handling.
- `pandas`: Tabular data loading, inspection, and analysis.
- `os`: File and directory path operations.
- `pathlib`: Robust cross-platform path management.
- `warnings`: Controlling warning output during data loading.

```python
import os
import warnings
from pathlib import Path

import numpy as np
import pandas as pd

from utils.data_loader import (
    dataset_summary,
    duplicate_report,
    feature_summary,
    load_dataset,
    missing_value_report,
    validate_dataset,
)

warnings.filterwarnings('ignore')
```

## Step 2: Load Dataset

The reusable `load_dataset` function supports CSV and Excel files and raises clear errors for missing files or unsupported formats.

```python
project_root = Path.cwd().resolve()
raw_dir = project_root / 'dataset' / 'raw'

flood_path = raw_dir / 'flood_dataset.xlsx'
rainfall_path = raw_dir / 'rainfall_in_india_1901_2015.xlsx'

try:
    flood_df = load_dataset(flood_path)
    print('Flood dataset loaded successfully.')
except Exception as exc:
    print(f'Flood dataset load failed: {exc}')
    flood_df = None

try:
    rainfall_df = load_dataset(rainfall_path)
    print('Reference rainfall dataset loaded successfully.')
except Exception as exc:
    print(f'Reference dataset load failed: {exc}')
    rainfall_df = None
```

## Step 3: Dataset Validation

Validation confirms whether the dataset exists, has rows and columns, and contains duplicate or null column names.

```python
if flood_df is not None:
    validation_summary = validate_dataset(flood_df, dataset_name='flood_dataset')
    validation_summary
else:
    print('Flood dataset is unavailable for validation.')
```

## Step 4: Explore Flood Dataset

The next cells inspect the structure and contents of the primary flood dataset.

```python
if flood_df is not None:
    display(flood_df.head())
    print('\nLast 5 Records:')
    display(flood_df.tail())
    print('\nRandom 5 Records:')
    display(flood_df.sample(5, random_state=42))
    print(f'\nShape: {flood_df.shape}')
    print(f'Rows: {flood_df.shape[0]}')
    print(f'Columns: {flood_df.shape[1]}')
    print('Column Names:', list(flood_df.columns))
    print('Column Index:', list(flood_df.index))
else:
    print('Flood dataset is unavailable for preview.')
```

## Step 5: Dataset Information

This step provides structural information, memory usage, and data types.

```python
if flood_df is not None:
    flood_df.info()
    print('Memory Usage (MB):', round(flood_df.memory_usage(deep=True).sum() / (1024 * 1024), 2))
    print('Data Types:')
    print(flood_df.dtypes)
    print('Numeric Columns:', list(flood_df.select_dtypes(include=[np.number]).columns))
    print('Categorical Columns:', list(flood_df.select_dtypes(exclude=[np.number]).columns))
else:
    print('Flood dataset is unavailable for info summary.')
```

## Step 6: Statistical Summary

This section generates descriptive statistics for the dataset.

```python
if flood_df is not None:
    print('Describe:')
    display(flood_df.describe())
    print('Describe (include all):')
    display(flood_df.describe(include='all'))
    print('Mean:')
    display(flood_df.mean(numeric_only=True))
    print('Median:')
    display(flood_df.median(numeric_only=True))
    print('Mode:')
    display(flood_df.mode(numeric_only=True).head())
    print('Variance:')
    display(flood_df.var(numeric_only=True))
    print('Standard Deviation:')
    display(flood_df.std(numeric_only=True))
    print('Minimum:')
    display(flood_df.min(numeric_only=True))
    print('Maximum:')
    display(flood_df.max(numeric_only=True))
    print('Percentiles:')
    display(flood_df.quantile([0.25, 0.5, 0.75, 0.95]))
    print('Skewness:')
    display(flood_df.skew(numeric_only=True))
    print('Kurtosis:')
    display(flood_df.kurt(numeric_only=True))
else:
    print('Flood dataset is unavailable for statistical summary.')
```

## Step 7: Target Variable Analysis

The target column is `flood`. Values are interpreted as `0 = No Flood` and `1 = Flood`.

```python
if flood_df is not None and 'flood' in flood_df.columns:
    target = flood_df['flood']
    print('Unique Values:', target.unique().tolist())
    print('Class Counts:')
    print(target.value_counts(dropna=False))
    print('Class Percentages:')
    print(target.value_counts(normalize=True).mul(100).round(2))
else:
    print('Target column is not available.')
```

## Step 8: Missing Values Analysis

Only missing values are detected here. No cleaning is performed.

```python
if flood_df is not None:
    missing_report = missing_value_report(flood_df)
    display(missing_report)
else:
    print('Flood dataset is unavailable for missing value analysis.')
```

## Step 9: Duplicate Analysis

The notebook detects duplicate rows without removing them.

```python
if flood_df is not None:
    duplicate_info = duplicate_report(flood_df)
    print('Duplicate Count:', duplicate_info['duplicate_count'])
    display(duplicate_info['duplicate_records'].head())
else:
    print('Flood dataset is unavailable for duplicate analysis.')
```

## Step 10: Feature Summary

A professional feature summary table is prepared for the dataset.

```python
if flood_df is not None:
    feature_table = feature_summary(flood_df)
    display(feature_table)
else:
    print('Flood dataset is unavailable for feature summary.')
```

## Step 11: Reference Dataset Exploration

The rainfall reference dataset provides historical context for rainfall patterns and is not used for training the flood prediction model.

```python
if rainfall_df is not None:
    print('Reference dataset shape:', rainfall_df.shape)
    print('Reference dataset columns:', list(rainfall_df.columns))
    print('Year Range:', rainfall_df['Year'].min(), 'to', rainfall_df['Year'].max())
    print('States:', sorted(rainfall_df['State'].dropna().unique().tolist())[:10], '...')
    print('Annual Rainfall Summary:')
    display(rainfall_df['Annual Rainfall'].describe())
    print('Seasonal Rainfall Summary:')
    display(rainfall_df['Seasonal Rainfall'].describe())
else:
    print('Reference dataset is unavailable for exploration.')
```

## Step 12: Generate Data Quality Report

A markdown report is prepared for future EDA and preprocessing phases.

```python
report_path = project_root / 'reports' / 'data_quality_report.md'
report_path.write_text(report_path.read_text(encoding='utf-8'), encoding='utf-8')
print(f'Report written to: {report_path}')
```
